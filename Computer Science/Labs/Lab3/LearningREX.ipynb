## Regular Expression Basics
.    =>Any character except newline<br>
a    =>The character a \n<br>
ab	 =>The string ab<br>
a|b	 =>a or b <br>
a*	 =>0 or more a's<br>
\	 =>Escapes a special character<br>

## Regular Expression Quantifiers
\*	   =>0 or more<br>
\+	   =>1 or more<br>
?	   =>0 or 1<br>
{2}	   =>Exactly 2<br>
{2, 5} =>Between 2 and 5
{2,}   =>2 or more<br>
(,5}   =>Up to 5<br>

## Regular Expression Groups
(...)  =>Capturing group<br>

## Regular Expression Assertions<br>
^	=>Start of string<br>
$	=>End of string<br>

## Regular Expression Character Classes<br>
\[ab-d]	=>One character of: a, b, c, d<br>
\[^ab-d]	=>One character except: a, b, c, d<br>
\[\b]	=>Backspace character<br>
\d	    =>One digit<br>
\D	    =>One non-digit<br>
\s	    =>One whitespace<br>
\S	    =>One non-whitespace<br>
\w	    =>One word character<br>
\W	    =>One non-word character<br>
