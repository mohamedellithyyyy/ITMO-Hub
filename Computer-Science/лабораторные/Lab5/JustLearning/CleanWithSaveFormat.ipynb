```python
from openpyxl import load_workbook
```

```python
wb = load_workbook("../lab5.xlsx")
ws = wb["Sheet2"]
```

```python
for row in ws.iter_rows(min_col=6, max_col=6, min_row=1, max_row=ws.max_row):
    for cell in row:
        cell.value = None 
```

```python
wb.save("../mainCleaned.xlsx")
```
