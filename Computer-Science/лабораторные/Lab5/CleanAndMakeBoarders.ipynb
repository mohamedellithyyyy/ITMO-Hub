```python
import pandas as pd
from openpyxl.styles import Border, Side
from openpyxl import load_workbook
```

```python
# Read file
df = pd.read_excel("lab5.xlsx", sheet_name="Sheet2", header=None)
```

```python
# Clean column instead of deleting
df[5] = None
```

```python
# Save temporary
temp_path = "output.xlsx"
df.to_excel(temp_path, index=False, header=False)
```

```python
# Reopen to apply borders
wb = load_workbook(temp_path)
ws = wb.active
```

```python
# Blue border
blue = Side(border_style="thin", color="0000FF")
border = Border(left=blue, right=blue, top=blue, bottom=blue)
```

```python
# Apply border except B1:C2
skip = {"B1", "B2", "C1", "C2"}

for row in ws.iter_rows(min_row=1, max_row=ws.max_row,
                        min_col=1, max_col=ws.max_column):
    for cell in row:
        if cell.coordinate not in skip:
            cell.border = border
```

```python
# Save result
wb.save("output_lab5.xlsx")
```
